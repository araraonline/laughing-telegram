# BetExplorer Teams


## General


### Amount of teams

There are 10161 different names for teams (raw).

### Team IDs

Teams IDs are consistent across championships. For example, take a look at these:

 - http://www.betexplorer.com/soccer/brazil/serie-a/teaminfo.php?team_id=xUT0Bp8o
 - http://www.betexplorer.com/soccer/brazil/campeonato-mineiro/teaminfo.php?team_id=xUT0Bp8o
 
Both pages refer to the same team (América MG), but in different leagues.

### No duplicates

There are no duplicate teams.

### Characters

The following characters are present:

     &'()-./0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz’

Examples:
- Bosnia & Herzegovina
- Bishop's Stortford
- Pirin Blagoevgrad (1922) \*
- Al-Sadd
- Argentina Ol.
- Bodo/Glimt
- Taifa Jang’ombe (Zan) \*\*


\* This team is unique because it contains a year  
\*\* This team is unique because it has the character `’`  
\*\*\* Parenthesis are only used for country abbreviations and "(Am)"

### Special tokens

##### Under-XX

Teams that are composed of under XX players are represent with a "UXX" at the end of the name. 

Examples: 

- Botafogo U20
- Rio Branco AC U20

##### Women

Teams that are made up of female players come with the token "W" at the end.

Examples: 

- Angola W
- Aruba W

##### Country Abbreviations

Country abbreviations for the teams are always of the form `(COU)`, where "COU" is a 3-lettered abbreviation for the country. In some places it is used, in some places it is not. Nevertheless, the team ID keeps consistent.

Examples: 

- AIK (Swe)
- Campinense (Bra)

##### (Am)

There are some teams that contain an "(Am)" in the name. We are not interested in any teams that have this.

### Tokens order

The tokens follow this given order:

    Team Name UXX W (Cty)

## Brazil

#### States

If the state is considered part of the name of the team, it comes prepended with a dash. Otherwise, it is just separated with a space.

Examples:
- 7 de Setembro (*no state*)
- Atletico-MG
- Audax RJ
